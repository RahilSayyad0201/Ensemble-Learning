```python
#import libraries
import numpy as np 
import pandas as pd
```

## read dataset

```python
df=pd.read_csv("breast_cancer.csv")
```

```python
df
```

```python
df.isnull().sum()## check null value
```

```python
df["diagnosis"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder #convert categorical variables into numerical variables.
le=LabelEncoder()
df["diagnosis"]=le.fit_transform(df["diagnosis"])
df["diagnosis"].unique()
```

```python
df.drop("id",axis=1,inplace=True)# drop the id column
```

```python
df["diagnosis"].value_counts()
```

## separate data in x (independant) and y(dependant) variable

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=11)
```

## fitting the model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rf=RandomForestClassifier(n_estimators=100)
```

```python
c=x.columns
```

```python
pd.DataFrame(Rf.feature_importances_,index=c)
```

## fit the model on traing set

```python
Rf.fit(x_train,y_train)
```

## predict the new observation

```python
y_pred=Rf.predict(x_test)
```

## check accuracy using avalable data(y_test) and predicted data (y_pred)

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix #confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report #classification_report
print(classification_report(y_test,y_pred))
```
