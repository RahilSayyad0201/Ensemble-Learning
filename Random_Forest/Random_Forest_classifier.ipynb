## import libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

## read data

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df.isnull().sum()
```

## Drop the column deck

```python
df.drop("deck",axis=1,inplace=True)
```

## Drop the column survived

```python
df.drop("survived",axis=1,inplace=True)
```

## fill the value in age column

```python
df[['age']] = df[['age']].mean()
```

## fill the value in embark_town column

```python
my_list = 'Southampton'
df['embark_town'] = df['embark_town'].fillna(my_list)
```

## fill the value in embarked column

```python
my_list = ['S','C','Q']
for i in my_list:
    df['embarked'] = df['embarked'].fillna(i)
```

```python
df #printing updated data
```

## Using LabelEncoder convert string value to new format(int)

```python
l=['sex','embarked','class','who','adult_male','embark_town','alive','alone']
for i in l:
    from sklearn.preprocessing import LabelEncoder
    le=LabelEncoder()
    le.fit(df[i])
    df[i]=le.transform(df[i])
```

```python
df # updated columns value
```
